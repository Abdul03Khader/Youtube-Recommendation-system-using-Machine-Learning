```python
#filling null values
df['description'] = df['description'].fillna(' ')
```

```python
df.isna().sum()
```

```python
#selecting important column
df = df[['title','description','tags']]
```

```python
#combining all
df['content'] = df['title'] + " " + df['description'] + " " +df['tags']
```

```python
#text cleaning
import re

def clean(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-z]', " " , text)
  return text
```

```python
df['content'] = df['content'].apply(clean)
```

```python
df[['title','content']].head()
```

```python
#TF -idf
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words = 'english',
    max_features=5000
)

#x variable
x = vectorizer.fit_transform(df['content'])
```

```python
#COSINE SIMILARITY
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(x)
```

```python
#function for reccommendation
def recommend(video_title):

  if video_title not in df['title'].values:
    print(f"'{video_title}' video not found")


  idx = df[df['title'] == video_title].index[0] # finds the row number (index) of that specific video
  score = list(enumerate(similarity[idx])) # list of scores of similar videos, enumatare it pairs scores with its original index
  score = sorted(score, key=lambda x: x[1], reverse=True) # It sorts the list from highest to lowest similar videos
  score = score[1:6] # first 5 similar videos

  video_indices = [i[0] for i in score]  # extracts the index numbers from your sorted list of pairs.
  return df['title'].iloc[video_indices] # Uses these index numbers to fetch the actual titles from the original DataFrame and returns them as a clean list.

```

```python
!pip install gradio
import gradio as gr

def gradio_recommend(video_title):
  recommendations = recommend(video_title)
  if recommendations.empty:
    return "No recommendations found or video title not in dataset."
  else:
    return "\n".join(recommendations.tolist())

iface = gr.Interface(
    fn=gradio_recommend,
    inputs=gr.Textbox(lines=2, placeholder="Enter a video title..."),
    outputs="text",
    title="YouTube Video Recommender",
    description="Enter a YouTube video title to get 5 similar video recommendations."
)

iface.launch()
```
